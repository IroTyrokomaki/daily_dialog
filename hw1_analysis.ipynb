```python
import re

file_path = "ijcnlp_dailydialog/train/dialogues_train.txt"

with open(file_path, "r", encoding="utf-8") as file:
    dialogues = [line.strip() for line in file.readlines()]

# Initialize lists to store user and system turns for each dialogue
all_user_turns_per_dialogue = []
all_system_turns_per_dialogue = []


def split_user_system_turns_single_string(dialogue):
    turns = dialogue.split('__eou__')

    # Remove empty turns
    turns = [turn.strip() for turn in turns if turn.strip()]

    # Separate user and system turns
    user_turns = turns[::2]
    system_turns = turns[1::2]

    # Remove extra space before punctuation in each turn
    user_turns = [re.sub(r'\s+([.,;!?])', r'\1', turn) for turn in user_turns]
    system_turns = [re.sub(r'\s+([.,;!?])', r'\1', turn) for turn in system_turns]

    return user_turns, system_turns

# Apply the splitting function to each dialogue
for dialogue in dialogues:
    user_turns, system_turns = split_user_system_turns_single_string(dialogue)
    all_user_turns_per_dialogue.append(user_turns)
    all_system_turns_per_dialogue.append(system_turns)

# Print or use the results
for i in range(len(all_user_turns_per_dialogue)):
    print(f"\nDialogue {i + 1}:")
    print("User Turns:")
    print(all_user_turns_per_dialogue[i])
    print("System Turns:")
    print(all_system_turns_per_dialogue[i])
```

### Total data length

```python
total_dialogues = len(all_user_turns_per_dialogue)
print(f"Total Dialogues: {total_dialogues}")
```

```python
# Total dialogues length for user/system
total_length_user = sum(len(turn.split()) for dialogue in all_user_turns_per_dialogue for turn in dialogue)
total_length_system = sum(len(turn.split()) for dialogue in all_system_turns_per_dialogue for turn in dialogue)

# Print the results
print(f"Total User Dialogue Length: {total_length_user}")
print(f"Total System Dialogue Length: {total_length_system}")
```

```python
# Total Word Length for user/system

import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

# Calculate total user and system words
total_user_words = sum(len(word_tokenize(turn)) for turns in all_user_turns_per_dialogue for turn in turns)
total_system_words = sum(len(word_tokenize(turn)) for turns in all_system_turns_per_dialogue for turn in turns)
total_words = total_user_words + total_system_words

print(f"\nTotal User Words: {total_user_words}")
print(f"Total System Words: {total_system_words}")
print(f"Total Words: {total_words}")
```

```python
# Total Sentences Length for user/system

from nltk.tokenize import sent_tokenize

# Calculate total user and system sentences
total_user_sentences = sum(len(sent_tokenize(turn)) for turns in all_user_turns_per_dialogue for turn in turns)
total_system_sentences = sum(len(sent_tokenize(turn)) for turns in all_system_turns_per_dialogue for turn in turns)
total_sentences = total_user_sentences + total_system_sentences

print(f"\nTotal User Sentences: {total_user_sentences}")
print(f"Total System Sentences: {total_system_sentences}")
print(f"Total Sentences: {total_sentences}")
```

```python
# Total Turns Length for user/system

# Calculate total user and system turns
total_user_turns = sum(len(turns) for turns in all_user_turns_per_dialogue)
total_system_turns = sum(len(turns) for turns in all_system_turns_per_dialogue)

print(f"Total User Turns: {total_user_turns}")
print(f"Total System Turns: {total_system_turns}")
```

### Mean data Length

```python
# Calculate average length for user and system dialogues
average_length_user = total_length_user / total_dialogues
average_length_system = total_length_system / total_dialogues

# Print the results
print(f"Average Length User Dialogue: {average_length_user:.1f}")
print(f"Average Length System Dialogue: {average_length_system:.1f}")
```

```python
# Mean leangth in terms of sentences

average_user_sentences = total_user_sentences / total_dialogues
average_system_sentences = total_system_sentences / total_dialogues

print(f"Average User Sentences per Dialogue: {average_user_sentences:.1f}")
print(f"Average System Sentences per Dialogue: {average_system_sentences:.1f}")
```

```python
# Mean leangth in terms of turns

average_user_turns = total_user_turns / total_dialogues
average_system_turns = total_system_turns / total_dialogues

average_speaker_turns_per_dialogue = average_user_turns + average_system_turns

print(f"Average User Turns per Dialogue: {average_user_turns:.1f}")
print(f"Average System Turns per Dialogue: {average_system_turns:.1f}")
print(f"\nAverage Speaker Turns Per Dialogue: {average_speaker_turns_per_dialogue:.1f}")
```

```python
# Mean leangth in terms of tokens(words)
# Average Tokens(words) Per Dialogue is written also on Basic Statistics in http://yanran.li/dailydialog.html

average_tokens_per_dialogue = sum(len(dialogue.split()) for dialogue in dialogues) / total_dialogues
print(f"Average Tokens Per Dialogue: {average_tokens_per_dialogue:.1f}")
```

```python
# Vocabulary size for user/system
from collections import Counter

all_user_words = [word for turns in all_user_turns_per_dialogue for turn in turns for word in turn.split()]
all_system_words = [word for turns in all_system_turns_per_dialogue for turn in turns for word in turn.split()]

# Calculate vocabulary size for user and system
vocabulary_size_user = len(Counter(all_user_words))
vocabulary_size_system = len(Counter(all_system_words))

print(f"Vocabulary Size User: {vocabulary_size_user}")
print(f"Vocabulary Size System: {vocabulary_size_system}")
```
